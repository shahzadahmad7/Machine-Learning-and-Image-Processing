```python
from skimage import io
```

```python
img=io.imread('mg.jpg')
io.imshow(img)
```

```python
from skimage import color
from skimage import data
from pylab import *
#Convert to hsv
mg_hsv=color.rgb2hsv(img)
figure(0)
io.imshow(mg_hsv)

#Convert back to rgb
mg_rgb=color.hsv2rgb(mg_hsv)
figure(1)
io.imshow(mg_rgb)
```

```python
#Convert to xyz
mg_xyz=color.rgb2xyz(img)
figure(0)
io.imshow(mg_xyz)
```

```python
#Convert to LAB
mg_lab=color.rgb2lab(img)
figure(1)
io.imshow(mg_lab)
io.imsave('ss.jpg', mg_lab)
```

```python
#Convert to grey
mg_grey=color.rgb2gray(img)
figure(2)
io.imshow(mg_grey)
```

```python
from skimage import draw
x,y=draw.line(0,0,1000,1000)
img[x,y]=0
io.imshow(img)
```

```python
# The circle function is used to draw a circle. In the following code, the first
# two arguments indicate the position of the circle inside the image; the last
# argument indicates the radius.

x,y=draw.circle_perimeter(300,300, 100)
img[x,y]=1
io.imshow(img)
```

```python
# The first six arguments in the following code define three points; the
# last argument defines the tension present in the line.

x,y=draw.bezier_curve(0,0, 500, 500, 1200, 900, 100)
img[x,y]=1
io.imshow(img)
```

```python
from skimage import exposure
gamma1=exposure.adjust_gamma(img, 0.5)
gamma2=exposure.adjust_gamma(img, 5)
figure(0)
io.imshow(gamma1)

figure(1)
io.imshow(gamma2)
```

**Rotating, Shifting, and Scaling Images**

Sometimes we may want to rotate an image or change its size. To do this,
we use the transform class in the skimage module. transform has two
functions: rotate and resize. rotate takes the degree of rotation as its
parameter; resize takes the new size as its parameter.

```python
from skimage.transform import resize, rotate
image_res=resize(img,(100,100))
figure(0)
io.imshow(image_res)
figure(1)
image_rot=rotate(img, 20)
io.imshow(image_rot)
```

 **structural similarity** is used to find the index that
indicate how much two images are similar.

```python
from skimage.metrics import structural_similarity as ssim
from skimage import io, img_as_float

# Load the images
img_original = io.imread('mg.jpg')
img_modified = io.imread('ss.jpg')

# Convert images to float format
img_original = img_as_float(img_original)
img_modified = img_as_float(img_modified)

# Calculate SSIM
ssim_original = ssim(img_original, img_original, data_range=img_original.max() - img_original.min(), multichannel=True)
ssim_different = ssim(img_original, img_modified, data_range=img_modified.max() - img_modified.min(), multichannel=True)

# Print the results
print("SSIM between original and original:", ssim_original)
print("SSIM between original and modified:", ssim_different)

```
